```python
import tensorflow as tf
```

```python
tensor_zero_d = tf.constant(4)
tensor_zero_d2 = tf.constant(4,dtype=tf.float32)
```

```python
print(tensor_zero_d)
print(tensor_zero_d2)
```

```python
tensor_one_d = tf.constant([2,0,-3])
print(tensor_one_d)
```

```python
tensor_one_d = tf.constant([2,0,-3.2])
print(tensor_one_d)
```

```python
tensor_3d = tf.constant([

[[1,2,0],
[3,5,-1]],

[[10,2,0],
[1,0,2]],

[[5,8,0],
[2,7,0]],

[[2,1,9],
[4,-3,32]]

])
print(tensor_3d)
```

```python
print(tensor_3d.shape)
```

we can stack several 3d tensors to create a 4d tensor.

```python

tensor_4d =tf.constant([

[[[1,2,0],
[3,5,-1]],

[[10,2,0],
[1,0,2]],

[[5,8,0],
[2,7,0]],

[[2,1,9],
[4,-3,32]]],

[[[1,2,0],
[3,5,-1]],

[[10,2,0],
[1,0,2]],

[[5,8,0],
[2,7,0]],

[[2,1,9],
[4,-3,32]]],

[[[1,2,0],
[3,5,-1]],

[[10,2,0],
[1,0,2]],

[[5,8,0],
[2,7,0]],

[[2,1,9],
[4,-3,32]]]


])

print(tensor_4d)
```

we could use tf.cast function to change the data type of the variable

```python
casted_tensor = tf.cast(tensor_zero_d,dtype=tf.float32)
print(casted_tensor)
```
